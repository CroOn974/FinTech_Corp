```python
import yfinance as yf

# Récupération des données sur le bitcoin
df_BTC = yf.download('BTC-EUR', start='2013-01-01')
```

```python
df_BTC_Cp = df_BTC.copy()
```

```python
# importation des bibliothèques nécessaires
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Ré-échantillonnage de la colonne 'Close' pour les données hebdomadaires et calcul de la moyenne
df_BTC_mean = df_BTC_Cp.resample('W').mean()

# Ajout d'une colonne 'Result' qui indique si le prix de fermeture de la semaine en cours est supérieur à celui de la semaine suivante
df_BTC_mean['Result'] = df_BTC_mean['Close'] > df_BTC_mean['Close'].shift(-1)
```

```python
df_BTC_mean
```

```python
# Réinitialisation de l'index du DataFrame df_BTC_mean
df_BTC_mean.reset_index(inplace=True)

# Contient les 3 dernières lignes de "df_BTC_mean"
df_BTC_last = df_BTC_mean.tail(3)
df_BTC_last = df_BTC_last.drop(['Date','Result'], axis=1)

# Enlève les 3 dernières lignes de "df_BTC_mean"
df_BTC_mean = df_BTC_mean[:-3]

# Stock la colonne 'Result' dans la variable 'y'
y = df_BTC_mean['Result']

# Stockage les colonne contenues dans le DataFrame df_BTC_mean dans la variable 'X' en enlevant les colonnes Date et Result
X = df_BTC_mean.drop(['Date','Result'], axis=1)
```

```python
df_BTC_mean
```

```python
df_BTC_last
```

```python
from sklearn.model_selection import train_test_split

# Séparer les données en ensemble d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
# Import de la classe KNeighborsClassifier de sklearn pour créer un modèle de classification basé sur les k plus proches voisins.
from sklearn.neighbors import KNeighborsClassifier

# Initialisation d'un modèle KNeighborsClassifier
model = KNeighborsClassifier()

# Entraînement du modèle sur les données d'entraînement
model.fit(X_train, y_train)

# Evaluation de la performance du modèle en utilisant les données de test
accuracy = model.score(X_test, y_test)
print("Accuracy: ", accuracy)
```

```python
def invest(model, dataframe):
    for row in dataframe.values:
        print(model.predict(row.reshape(1,-1)))
        print(model.predict_proba(row.reshape(1,-1)))
```

```python
invest(model, df_BTC_last)
```

```python
# 2. Bagging
```

```python
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier
# Initialisation d'un modèle de classification combiné appelé BaggingClassifier
# qui utilise un classificateur basé sur les plus proches voisins (KNeighborsClassifier) comme classificateur de base
# et 100 arbres de décision pour les combiner
model_Bagging = BaggingClassifier(base_estimator=KNeighborsClassifier(),
                         n_estimators=100)

# Entraînement du modèle sur les données d'entraînement
model_Bagging.fit(X_train, y_train)

# Evaluation de la performance du modèle en utilisant les données de test
accuracy = model_Bagging.score(X_test, y_test)
print("Accuracy: ", accuracy)
```

```python
invest(model_Bagging, df_BTC_last)
```

```python
# 3. Boosting
```

```python
# Import de la classe AdaBoostClassifier de sklearn.ensemble
from sklearn.ensemble import AdaBoostClassifier

# Initialisation d'un objet model avec AdaBoostClassifier
# n_estimators est utilisé pour spécifier le nombre d'estimateurs (ou arbres de décision) utilisés dans l'algorithme
model_Boosting = AdaBoostClassifier(n_estimators=100)

# Entraînement du modèle sur les données d'entraînement
model_Boosting.fit(X_train, y_train)

# Evaluation de la performance du modèle en utilisant les données de test
accuracy = model_Boosting.score(X_test, y_test)
print("Accuracy: ", accuracy)
```

```python
invest(model_Boosting, df_BTC_last)
```
